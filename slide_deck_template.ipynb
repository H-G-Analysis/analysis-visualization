## Investigation Overview

> Our Investigation is concerned with comparing Different criteria with trip duration. Our interest is in the relationship between user_type, gender, age and trip duration.

## Dataset Overview

> Ford GoBike dataset include 183412 record for trips data covering San Francisco Bay area. with 16 Features like start and end station, trip duration, bike identifier and user data like user type, gender and birthyear. We added Column for Age. After cleaning data we ended with 174708 row and 17 column. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike_df = pd.read_csv('bike_clean.csv')
```

## 1. Gender and Subscription proportions

- Male users are greater than Female users.Male users are 74.6% of users, Female users are 23.3% and others are 2.1%.
- Number of subscriber is much greater than number of customers. 90.5% of users are subscribers. 9.5% of users are customers.

```python
plt.figure(figsize=[20,5])
plt.subplot(1,3,1)
#Create pie chart for users gender
gender_count = bike_df.member_gender.value_counts()
plt.pie(gender_count, labels=gender_count.index,startangle=90, counterclock=False, autopct='%1.1f%%',shadow=True,explode=(0.1, 0.1, 0.1) )
#Apply scaling to the plot to be equal on both the x- and y-axes. 
plt.axis('square')
plt.title('Gneder Proportion')

plt.subplot(1,3,2)
#Create pie chart for users type
user_type_count = bike_df.user_type.value_counts()
plt.pie(user_type_count, labels=user_type_count.index,startangle=90, counterclock=False, autopct='%1.1f%%',shadow=True,explode=(0.1, 0.1) )
#Apply scaling to the plot to be equal on both the x- and y-axes. 
plt.axis('square')
plt.title('Subscripion Proportion')

#Bar plot shows relation between gender and subscribtion
plt.subplot(1,3,3)
sb.countplot(data = bike_df, x = 'member_gender', hue = 'user_type')
plt.title('Proportion of subscription for each Gender')
```

## 2. Duration, Birthyear and User type.

- Customer trip duration in almost all birthyears was greater than subscribers trip duration

```python
#Point plot shows the average trip duration for every birthyear from different user types
plt.figure(figsize=[20,5])
sb.pointplot(data=bike_df, x='member_birth_year', y='duration_sec',hue='user_type', errwidth=0)
plt.xticks(rotation=90)
plt.title('User-Type and Birthyear Vs. Duration')
plt.xlabel('Birth Year')
plt.ylabel('Duration (sec)')
plt.legend(loc='upper right', title='User Type')
```

## 3. Gender and Birthyear Vs. Trip Duration
- Eldery Females and other gender who were born from 1920 to 1949 have slight number of trips.
- Eldery males have two apikes in trip duration. with average trip duration about 2000 seconds.
- others have number of spikes in the middle age.



```python
#Point plot shows the average trip duration for every birthyear with different gender
plt.figure(figsize=[20,5])
sb.pointplot(data=bike_df, x='member_birth_year', y='duration_sec',hue='member_gender', errwidth=0)
plt.xticks(rotation=90)
plt.title('Gender and Birthyear Vs. Duration')
plt.xlabel('Birth Year')
plt.ylabel('Duration (sec)')
plt.legend(loc='upper right',title='Gender')
```

## Conclusion
